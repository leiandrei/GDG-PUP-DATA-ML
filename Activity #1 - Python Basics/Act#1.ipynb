#### **Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### **Menu Dict**

```python
menu = {
    "Lumpia Shanghai": 500, "Pinoy Spaghetti": 800, "Creamy Carbonara": 900,
    "Chicken Bucket": 750, "Lechon Belly": 3500, "Pork BBQ Platter": 600,
    "Buko Salad": 450, "Mango Graham": 400, "Pichi-Pichi": 250,
    "Rice Tray": 300, "Soft Drinks (Case)": 500, "Ice Cream Tub": 350,
    "Red Ribbon Cake": 900, "Fruit Platter": 300
}
```

### **FUNCTIONS**

```python
def show_menu(menu):
    
    print("MENU")

    for items, price in menu.items():
        print(f'{items} - PHP {price} ')
    
    print("\n")

def add_to_menu(menu):

    amount = 0
    orders = {}

    while True:

        order = input("Enter food to order (or type 'done'): ")

        if order == 'done':
            print(f'\nTOTAL SPENT: PHP {amount}')
            print('PERFECT PARTY! Good job, Sparky!')
            return amount, orders

        if order in menu:   
            print(f'Added {order} (PHP {menu[order]})')
            orders[order] = orders.get(order, 0) + menu[order]
            amount += menu[order]
        
 
def pie_chart(orders):

    plt.figure(figsize=(4, 4))
    plt.title('Final Party Menu Breakdown', fontweight="bold")
    plt.pie(orders.values(), labels=orders.keys(), autopct='%1.1f%%')
    plt.axis('equal')
    plt.show()

def bar_chart(amount, max_budget):

    plt.figure(figsize=(4, 4))
    plt.title('Budget vs Spending Comparison', fontweight="bold")
    plt.bar(['Budget', 'Total Spent'], [max_budget, amount])
    plt.show()

def budget_comp(orders, max_budget, min_budget):

    total_spent = sum(orders.values())

    if total_spent > max_budget:
        print("Over Budget")
        bar_chart(total_spent, max_budget)

    elif total_spent < min_budget:
        print('Under Budget')
        bar_chart(total_spent, max_budget)
    
    else:
        print('Successful')
        pie_chart(orders)
```

```python
if __name__ == "__main__":
    max_budget = 8000
    min_budget = 5000
    show_menu(menu)
    amount, orders = add_to_menu(menu)
    budget_comp(orders, max_budget, min_budget)
```
